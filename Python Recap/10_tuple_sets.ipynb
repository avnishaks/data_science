```python
"""
Set : unordered, mutable, and unique collection of elements. It does not allow duplicate values.

1. subset 
2. superset
3. difference 
4. symmetric_difference
5. union
6. intersection

"""
```

```python
s1 ={1,2,3,4,5,6}
```

```python
s2={4,5,6,7,8,9,10}
```

```python
s1.union(s2)
```

```python
s1.intersection(s2)
```

```python
s1.difference(s2)
```

```python
s2.difference(s1)
```

```python
s3={7,8,9}
```

```python
s3.issubset(s2)
```

```python
s2.issuperset(s3)
```

```python
s3.issuperset(s2)
```

```python
s1.symmetric_difference(s2)
```

```python
s2.symmetric_difference(s1)
```
