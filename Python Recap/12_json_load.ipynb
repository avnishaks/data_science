```python
"""
JSON Module :- 

1. json.dumps(obj) - Convert Python Object to JSON String
2. json.dump(obj,file) - Writes JSON to file 
3. json.loads(obj) - Convert JSON String to Python Object
4. json.load(obj,file) - Read JSON from file
"""
```

```python
import json
```

```python
obj = {
  "name": "Alice",
  "company": "IBM",
  "salary": 45000
}
```

```python
obj
```

```python
type(obj)
```

```python
s=json.dumps(obj) # Convert Python Object to JSON String
```

```python
type(s)
```

```python
d= json.loads(s) # Convert JSON String to Python Object
```

```python
type(d)
```

```python
with open("data.json", "w") as obj:
    json.dump(s, obj)
```

```python
with open("data.json", "r") as obj:
    loaded_file = json.load(obj)
```
